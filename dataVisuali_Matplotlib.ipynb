3. Histogram

A histogram chart in pandas (and more broadly in data analysis) is a graphical representation of the distribution of numerical data. It divides the data into intervals (or bins) and shows how many data points fall into each interval.

5. PIE PLOT


A pie plot (or pie chart) is a circular statistical graphic divided into slices to illustrate numerical proportions. Each slice represents a category's contribution to the whole, making it useful for showing percentage or proportional data.

```python
country_population = [450,730,900,450,800]
country_name = ['Japan','bangladesh','india','nepal','china']

plt.pie(country_population,labels= country_name)
plt.show()
```

autopct="%1.2f%%" formats the percentage labels on a pie chart to display with two decimal places, providing more precision than the default formatting.

```python
plt.pie(country_population,labels= country_name,autopct="%1.2f%%")
plt.show()
```

```python
#The explode parameter in pie charts is used to offset specific slices from the center, making them stand out visually. 

country_population = [450,730,900,450,800]
country_name = ['Japan','bangladesh','india','nepal','china']
explosion = [0.3,0,0,0,0.2]
plt.pie(country_population,labels= country_name,autopct="%2.2f%%",explode=explosion)
plt.show()
```

```python
#wedgeprops to customize the edge color and line width of the pie chart slices:

country_population = [450,730,900,450,800]
country_name = ['Japan','bangladesh','india','nepal','china']
explosion = [0.3,0,0,0,0.2]
plt.pie(country_population,labels= country_name,autopct="%2.2f%%",explode=explosion,wedgeprops={'ec':'black','lw':3})
plt.show()
```

```python
plt.figure(figsize=(6,8)) #------> width & height
country_population = [450,730,900,450,800]
country_name = ['Japan','bangladesh','india','nepal','china']
explosion = [0.3,0,0,0,0.2]
plt.pie(country_population,labels= country_name,autopct="%2.2f%%",explode=explosion,wedgeprops={'ec':'black','lw':3},shadow=True)
plt.savefig('country_pieplot')
plt.show()
```

Custom Function to Calculate Mean and Median

```python
import statistics

def calculate_mean_and_median(values):
    # Calculate mean
    mean = sum(values) / len(values)
    
    # Calculate median using statistics module
    median = statistics.median(values)
    
    return mean, median

# List of numbers
numbers = [10, 20, 21, 23, 26, 35, 31, 27]

# Calculate mean and median
mean, median = calculate_mean_and_median(numbers)

print("The mean is:", mean)
print("The median is:", median)
```
